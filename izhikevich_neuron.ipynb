### **Parametri di tempo**:

```python
tmax = 1000. # 1 secondo di attività espresso in ms (il " . " è per indicare che è un numero con la virgola. Su Python è importante)
dt = 0.5 # con uno step di attività di 0.5 ms per volta
```

Il modello di Izhikevich è fortemente parametrizzato. Non è possibile variare di troppo i parametri altrimenti si rischia di uscire dal comportamento neuronale modellizzato in queste equazioni.

Schermata 2021-10-08 alle 11.28.09.png

## **Scegli il modello neuronale da simulare:**

```python
a = 0.02
b = 0.2
c = -65
d = 8.
```

Puoi scegliere gli altri in base al modello che vuoi emulare (vedi figura sopra).


Adesso è il momento dei parametri di input:
- Corrente,
- intervallo di tempo in cui si interverrà con la stimolazione elettrica.

##**Scegli un valore di corrente da -100 a 100:**

```python
Iapp = 10
tr = array([200.,700])/dt
```

Ora si deve impostare il bacino di memoria informatica che si occuperà di ospitare i dati della simulazione (un grafico vuoto da riempire):

```python
T = int(ceil(tmax/dt))
v = zeros(T)
u = zeros(T)
v[0] = -70 # Si fissa il potenziale di riposo
u[0] = -14 # Fissata la condizione di riposo
```

In questa parte di codice si popolano i valori (inizialmente vuoti del grafico) con quelli che verranno ricavati dalle due equazioni del modello.

```python
for t in arange(T-1):
  if t > tr[0] and t < tr[1]:
    I = Iapp
  else:
    I = 0
  # PRIMO SET DI EQUAZIONI
  if v[t]<35:
    dv = (0.04*v[t]+5)*v[t]+140-u[t]
    v[t+1] = v[t] + (dv+I)*dt
  # SECONDO SET DI EQUAZIONI
    du = a*(b*v[t]-u[t])
    u[t+1] = u[t] + dt*du
  else:
    v[t] = 35
    v[t+1] = c
    u[t+1] = u[t]+d
```

## **Plot del grafico in base al modello scelto.**

```python
figure(figsize=(15,8), dpi=100)
tvec = arange(0.,tmax,dt)
plot(tvec,v,'b',label='Voltage Trace')
xlabel('Time [ms]')
ylabel('Membrane voltage [mV]')
title("""SIMULAZIONE DI UN SINGOLO NEURONE""")
show()
```
